```python
# read file
df = pd.read_csv("data/HR_comma_sep.csv")
df.head()
```

## Preprocessing

```python
df_str = df

# bins' values (labels)
l_satisfaction_level = ['very_low', 'low', 'medium', 'good', 'excellent']
l_last_evaluation = [str(el) + '_LAST_E' for el in range(4)] # [0-2) or [2-4) or [4,7), [7,] months ago
l_number_project = [str(i) + '_N_PROJECT' for i in range(df['number_project'].max() + 1)]
work_accident_map = {0:'NO_A', 1:'YES_WA'}
left_map = {0:'NO_LEFT', 1:'YES_LEFT'}
promotion_map = {0:'NO_PROMO', 1:'YES_PROMO'}

df_str['satisfaction_level'] = pd.cut(df['satisfaction_level'], bins=5, right=False, labels=l_satisfaction_level)
df_str['last_evaluation'] = pd.cut(df['last_evaluation'], bins=4, right=False, labels=l_last_evaluation)
df_str['number_project'] = [str(el) + '_NP' for el in df['number_project']]
df_str['Work_accident'] = df['Work_accident'].map(work_accident_map).astype(str)
df_str['left'] = df['left'].map(left_map).astype(str)
df_str['promotion_last_5years'] = df['promotion_last_5years'].map(promotion_map).astype(str)
```
